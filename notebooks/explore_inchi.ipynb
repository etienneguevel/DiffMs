```python
import pandas as pd

df_combined = pd.read_csv(
    'data/fp2mol/combined/preprocessed/combined_train.csv'
)

df_combined.head()
```

```python
from rdkit import Chem
from rdkit.Chem import Draw

# InChI string
inchi = df_combined['inchi'].iloc[15]

# Convert InChI to molecule
mol = Chem.MolFromInchi(inchi)

# Draw 2D structure
img = Draw.MolToImage(mol, size=(400, 400))
img
```
